```python
#chi-square test 1 for different features: transactionAmount
from scipy.stats import chi2_contingency 

table = pd.crosstab(df_unido['Fraud'], df['transactionAmount'])

# Realizando el test Chi-cuadrado
chi2, p, dof, expected = chi2_contingency(table)

print(f"Chi2 Statistic: {chi2}, p-value: {p}")

#different payments methods strong correlation with fraud as chi is high and p-value is low 
```

```python
#chi-square test 2 for different features: 'No_Payments' 

table2 = pd.crosstab(df_unido['Fraud'], df_unido['No_Payments'])

# Realizar el test Chi-cuadrado
chi2, p, dof, expected = chi2_contingency(table2)

print(f"Chi2 Statistic: {chi2}, p-value: {p}")

#the relation between device and probability of fraud is very high
```

```python
#chi-square test 3 for different features: 'No_Orders'

table3 = pd.crosstab(df_unido['Fraud'], df_unido['No_Orders'])

# Realizar el test Chi-cuadrado
chi2, p, dof, expected = chi2_contingency(table3)

print(f"Chi2 Statistic: {chi2}, p-value: {p}")

#frequancy or amount of orders in a shorter period of time have an impact in fraudulent transsactions
```

```python
###1. MODIFY DF BY DROPING COLUMNS NNOT INTERESTION FOR TEST

# List of columns to drop
columns_to_drop = [
    "transactionId", "orderId", "paymentMethodId", "customerEmail",
    "customerPhone", "customerDevice", "customerIPAddress", 
    "customerBillingAddress"
]

# Drop columns
df_ML = df_unido.drop(columns=columns_to_drop)
df_ML.head(3)
```

```python
#2. CONVERT IN CATEGORICAL NON CATEGORICAL VALUES

#  One-Hot Encoding
df_encoded = pd.get_dummies(df_ML, columns=['paymentMethodProvider', 'orderState', 'paymentMethodType'])
print(df_encoded.isnull().sum())
```

```python
#3. EMPTY VALUES and replace by values below (True, 0 y True)

df_encoded['paymentMethodRegistrationFailure'].fillna(True, inplace=True)
df_encoded['transactionAmount'].fillna(0, inplace=True)
df_encoded['transactionFailed'].fillna(True, inplace=True)
print(df_encoded.isnull().sum())
```

```python
#4. DATA SPLIT FOR THE TEST

from sklearn.model_selection import train_test_split

features = df_encoded.drop('Fraud', axis=1)  # All columns except 'Fraud'
target = df_encoded['Fraud']  #Defining the target
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
(###El número 42 a menudo se elige como un valor para random_state (u otras semillas aleatorias) por su referencia cultural 
###como "la respuesta a la pregunta fundamental de la vida, el universo y todo lo demás", según la novela "La guía del 
###autoestopista galáctico" de Douglas Adams)
```

```python
df_encoded.head(1)
```

```python
#5.TRAIN THE MODEL
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

model = Pipeline([
    ('classifier', LogisticRegression())
])

model.fit(X_train, y_train)
```

```python
###try on fine - tune 

from sklearn.model_selection import GridSearchCV

param_grid = {
    'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100],
    'classifier__penalty': ['l1', 'l2']
}

grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

print(Best gourp of parameters:", grid_search.best_params_)
```

```python
# Adjusting model with best parameters
model.set_params(classifier__C=1, classifier__penalty='l2')
model.fit(X_train, y_train)
```

```python
###6.EVALUATING THE MODEL

from sklearn.metrics import classification_report, confusion_matrix

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()


scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)
```

```python

y_pred = knn.predict(X_test)


print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Exactitud (Accuracy):", accuracy_score(y_test, y_pred))
```
